```python
irr = pd.read_csv('data/clean_wb/irr.csv', sep='\t', usecols=['wb_title','wb_body'])
ele = pd.read_csv('data/clean_wb/ele.csv', sep='\t', usecols=['wb_title','wb_body'])
miz = pd.read_csv('data/clean_wb/miz.csv', sep='\t', usecols=['wb_title','wb_body'])
voi = pd.read_csv('data/clean_wb/voi.csv', sep='\t', usecols=['wb_title','wb_body'])
dvb = pd.read_csv('data/clean_wb/dvb.csv', sep='\t', usecols=['wb_title','wb_body'])

irr['site'] = 'irr'
ele['site'] = 'ele'
miz['site'] = 'miz'
voi['site'] = 'voi'
dvb['site'] = 'dvb'

df = pd.concat([irr,ele,miz,voi,dvb])
print(df.shape)
display(df.head())
```

## Most Commonly Used Words

```python
from collections import Counter
from string import punctuation as en_punc
```

```python
with open('aux/stop-words.txt') as f:
    STOP_WORDS = [l.strip() for l in f]

def tokenize(text): # Assumes the text is already word-segmented and is space-separated
    return text.split()

def prepare_text_for_lda(text): # same as lda
    text = str(text).replace('။', '').replace('၊', '')
    text = text.translate(str.maketrans('', '', en_punc))
    tokens = tokenize(text)
#     tokens = [token for token in tokens if len(token) > 4]
    tokens = [token for token in tokens if token not in STOP_WORDS]
#     tokens = [get_lemma(token) for token in tokens]
    return tokens
```

```python
def word_counter(text_list):
    text_list = [prepare_text_for_lda(t) for t in text_list]
    words = [a for b in text_list for a in b]
    words = [w for w in words if w not in STOP_WORDS]
    countdf = pd.DataFrame.from_dict(
        Counter(words),
        orient='index')
    countdf.columns = ['frq']
    countdf.sort_values('frq', ascending=False, inplace=True)
    return countdf
```

### Words in Article Titles

```python
N = 10
irr_words = word_counter(irr.wb_title)
ele_words = word_counter(ele.wb_title)
miz_words = word_counter(miz.wb_title)
voi_words = word_counter(voi.wb_title)
dvb_words = word_counter(dvb.wb_title)
print('irr')
display(irr_words.head(N))
print('ele')
display(ele_words.head(N))
print('miz')
display(miz_words.head(N))
print('voi')
display(voi_words.head(N))
print('dvb')
display(dvb_words.head(N))
```
